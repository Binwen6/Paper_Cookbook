```python
import matplotlib.pyplot as plt
from plotnine import *
import pandas as pd
import numpy as np
```

```python
# 假设 df 是一个 DataFrame，包含 x_var, y_var 和 z_var 列
# 这里我们创建一个简单的示例 DataFrame
np.random.seed(0)
df = pd.DataFrame({
    'x_var': np.repeat(list('ABCD'), 4),
    'y_var': np.tile(list('abcd'), 4),
    'z_var': np.random.rand(16)
})

# 使用 plotnine 绘制热图（这里更像是瓦片图）
(ggplot(df, aes(x='x_var', y='y_var', fill='z_var'))
 + geom_tile()
 + scale_fill_gradient(low='white', high='steelblue')  # 设置颜色渐变
 #+ theme_minimal()  # 使用简约主题
)
```

```python
# 假设 df 是一个 DataFrame，但它本身不直接包含热图所需的数据
# 我们需要有一个二维数组来表示 z_var 的值，以及对应的行和列标签

# 创建一个示例 DataFrame
np.random.seed(0)
df_raw = pd.DataFrame({
    'x_var': np.repeat(list('ABCD'), 4),
    'y_var': np.tile(list('abcd'), 4),
    'z_var': np.random.rand(16)
})

# 使用 pivot 方法来创建一个透视表（pivot table）
df_pivot = df_raw.pivot(index='x_var', columns='y_var', values='z_var')

# 获取行数和列数
nrows, ncols = df_pivot.shape

# 使用 matplotlib 绘制热图
plt.imshow(df_pivot.values, cmap='hot', interpolation='nearest')
plt.colorbar()  # 添加颜色条
plt.xticks(range(ncols), df_pivot.columns)  # 设置 x 轴标签
plt.yticks(range(nrows), df_pivot.index)  # 设置 y 轴标签
plt.title('Heatmap with Matplotlib')  # 设置图表标题
plt.show()  # 显示图表
```

```python
# 假设 df 是你的 DataFrame，x_var 和 y_var 是你想要绘制小提琴图的列名
np.random.seed(0)  # 为了示例的可重复性
df = pd.DataFrame({
    'x_var': np.repeat(list('ABC'), 100),
    'y_var': np.concatenate([np.random.normal(i, size=100) for i in range(1, 4)])
})

# 根据 x_var 的值将数据分组并创建列表，每个列表包含一个组的数据
violin_parts = [df[df['x_var'] == group]['y_var'].values for group in df['x_var'].unique()]

# 使用 matplotlib 绘制小提琴图
fig, ax = plt.subplots()
ax.violinplot(violin_parts, showmedians=True)
ax.set_xticks(range(1, len(violin_parts) + 1))
ax.set_xticklabels(df['x_var'].unique())
plt.show()
```

#### 在 ggplot2 中，为了创建饼图（pie chart），通常不使用 geom_bar() 结合 coord_polar()，因为这种方式更适合创建极坐标下的柱状图（bar plot in polar coordinates），而不是真正的饼图。对于饼图，我们可以使用 coord_polar() 搭配 geom_bar()，但需要将 width 设置为1来模拟饼图，同时设置 position 为 "fill" 以确保柱子堆叠起来形成一个完整的圆。

#### 然而，由于 ggplot2 没有直接的饼图几何对象，使用 coord_polar() 并不是标准做法。在 R 的 ggplot2 中，我们通常会使用其他包（如 ggplot2 的扩展包 ggplot2::waffle 或 ggplot2 的替代包 ggridges 或 ggforce）来创建饼图，但 plotnine（Python 的 ggplot2 实现）可能不支持这些高级功能。

```python
# 假设 df 是你的 DataFrame，x_var 是标签列，y_var 是值列（通常用于饼图的百分比）
# 注意：y_var 的值应该表示每个部分的相对大小或百分比，并且它们的和应该接近1（或100%）
np.random.seed(0)  # 为了示例的可重复性
df = pd.DataFrame({
    'x_var': ['A', 'B', 'C', 'D'],
    'y_var': np.random.rand(4)  # 这里只是随机值，通常需要是归一化的值
})
df['y_var'] = df['y_var'] / df['y_var'].sum() * 100  # 归一化到百分比

# 使用 matplotlib 绘制饼图
plt.pie(df['y_var'], labels=df['x_var'], autopct='%1.1f%%')
plt.axis('equal')  # 确保饼图是圆的，而不是椭圆的
plt.show()
```

```python
# 假设 df 是你的 DataFrame，x_var, y_var, z_var 是列名
np.random.seed(0)
df = pd.DataFrame({
    'x_var': list(range(10)),
    'y_var': np.cumsum(np.random.rand(10)),  # 示例数据，累加随机数
    'z_var': list('AABBCCDDEE')  # 分组变量
})

# 使用 plotnine 绘制面积图
(ggplot(df, aes(x='x_var', y='y_var', fill='z_var'))
 + geom_area(alpha=0.7)  # 设置透明度以区分重叠区域
 # + theme_minimal()  # 使用简约主题
)
```

```python
# 假设有5个分组，每个分组有10个数据点
np.random.seed(0)
group_data = {group: np.random.rand(10).cumsum() for group in 'ABCDE'}

# 创建一个包含所有分组数据的 DataFrame
df_list = []
for group, values in group_data.items():
    df_group = pd.DataFrame({'x_var': list(range(10)), 'y_var': values, 'z_var': [group] * 10})
    df_list.append(df_group)

# 将列表中的 DataFrame 合并为一个 DataFrame
df = pd.concat(df_list, ignore_index=True)

# 现在 df 是长格式的，可以直接用于 stackplot
plt.stackplot(df['x_var'], df.groupby('z_var')['y_var'].cumsum(), labels=df['z_var'].unique())
plt.legend(title='z_var')
plt.xlabel('x_var')
plt.ylabel('y_var')
plt.title('Area Plot with Matplotlib')
plt.show()
```

```python
# 假设 df 是您的 DataFrame，并且它包含 'x_var' 和 'y_var' 列
np.random.seed(0)
df = pd.DataFrame({
    'x_var': np.random.rand(100),
    'y_var': 2 * np.random.rand(100) + 3  # 假设有一个简单的线性关系 y = 2x + 3
})

# 使用 matplotlib 绘制散点图
plt.scatter(df['x_var'], df['y_var'], label='Data Points')

# 使用 np.polyfit 拟合线性回归模型，并绘制回归线
slope, intercept = np.polyfit(df['x_var'], df['y_var'], 1)
regression_line = slope * df['x_var'] + intercept
plt.plot(df['x_var'], regression_line, color='red', label='Regression Line')

# 添加图例和标题
plt.legend()
plt.title('Scatter Plot with Regression Line')
plt.show()
```
